# Merge historical_data and fear_greed_index 

```python
new_data = pd.merge(his_data,fear_data[['date','classification','value']],on='date',how='left')
```

```python
new_data
```

# Performance_summary                               

```python
group_by = new_data.groupby('classification').agg({'Closed PnL':'mean','Size USD':'sum','Execution Price':'mean',
                                                   'Account':'count'}).reset_index()
                                                   
```

```python
group_by
```

# Performance_Visualization

```python
plt.figure(figsize=(8,6))
sns.barplot(x='classification',y='Closed PnL',data=group_by,color='red')
plt.xlabel('Classification')
plt.ylabel('Average closed panel')
plt.show()           
```

```python
group_by['Closed PnL'].max()
```

# Key insight
   1.Extreme Greed is the highest aveage close panel.
